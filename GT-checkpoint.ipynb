```python
import statsmodels.api as sm
import numpy as np

X = df.drop('output', axis=1)  
y = df['output']  

X_const = sm.add_constant(X)

model = sm.Logit(y, X_const)
result = model.fit()

coefficients = result.params[1:]  
p_values = result.pvalues[1:]  

viz_df = pd.DataFrame({
    'Factors': coefficients.index,
    'Coefficients': coefficients.values,
    'P-Values': p_values.values
})

viz_df['Significance'] = np.where(viz_df['P-Values'] <= 0.05, 'Significant', 'Not Significant')

plt.figure(figsize=(12, 8))
sns.barplot(x='Coefficients', y='Factors', data=viz_df, hue='Significance', dodge=False, palette='viridis')
plt.title('Coefficients of Logistic Regression Model')
plt.xlabel('Coefficient Value')
plt.ylabel('Factors')
plt.axvline(x=0, color='grey', linestyle='--')
plt.legend(title='Statistical Significance', loc='lower right')

plt.show()

viz_df
```

```python
numeric_list=["age","trtbps","chol","thalachh","oldpeak","output"]
numeric_list_new=numeric_list[:-1]
df_numeric=df.loc[:, numeric_list_new]

for i in numeric_list_new:
    
    #IQR
    Q1=np.percentile(df.loc[:,i],25)
    Q3=np.percentile(df.loc[:,i],75)
    IQR=Q3-Q1
    print("Old shape: ",df.loc[:,i].shape)
    
    # upper bound
    upper=np.where(df.loc[:,i]>=(Q3+2.5*IQR))
    
    # lower bound
    lower=np.where(df.loc[:,i]<=(Q1-2.5*IQR))
    
    print("{} -- {}".format(upper,lower))
    
    try:
        df.drop(upper[0], inplace=True)
    except: print("KeyError: {} not found in axis".format(upper[0]))
    
    try:
        df.drop(lower[0], inplace=True)
    except: print("KeyError: {} not found in axis".format(lower[0]))
        
    print("New shape: ",df.shape)
df
```

```python
import xgboost as xgb
import numpy as np
import re
from sklearn.model_selection import StratifiedKFold,GridSearchCV
from sklearn.metrics import accuracy_score,roc_auc_score,log_loss,classification_report, confusion_matrix,precision_score,recall_score,f1_score
x=df[['age','sex','cp','trtbps','chol','fbs','restecg','thalachh'
     ,'exng','oldpeak','slp','caa','thall']]
# x=df[['sex','cp','thalachh','exng','oldpeak','caa','thall']]
y=df[['output']]
# X=df1.drop(["output"],axis=1)
# y=df1[["output"]]
from sklearn.model_selection import train_test_split
x_train, x_validation, y_train, y_validation= train_test_split(x, y, test_size=0.25, random_state=2023)
```

```python
from xgboost import XGBClassifier
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# y_train = le.fit_transform(y_train)
xg=XGBClassifier(random_state=2022)
xg.fit(x_train,y_train)
y_pred = xg.predict(x_validation)
#Xgbc_auc=roc_auc_score(y_validation,y_pred,multi_class='ovr') 
# print('Xgbc_auc:',Xgbc_auc)
# logloss=log_loss(y_validation,y_pred)
# print(logloss)
print(confusion_matrix(y_validation,y_pred))
accuracy_score1 = accuracy_score(y_validation,y_pred)
print("accuracy_score:",accuracy_score1)
precision_score1 = precision_score(y_validation,y_pred)   #average{‘micro’,‘samples’,‘weighted’,‘macro’} or None, default=’macro’
print("precision_score:",precision_score1)
recall_score1 = recall_score(y_validation,y_pred)
print("recall_score:",recall_score1)
f1score = f1_score(y_validation,y_pred)
print("f1_score:",f1score)
```
